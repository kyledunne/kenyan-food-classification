The dataset consists of **8,174 images in 13 Kenyan food type classes**.
Sample images of the KenyanFood13 dataset and the number of images in
each of the classes are shown below:

The dataset is split into a public train set and a private test set.

1.  You have to use the **train.csv** file for training. You can split
    the public subset into train and validation sets yourself.
2.  **test.csv** contains the ID of the images chosen for the Test set.
3.  After you have finished training, generate a **`.csv`** file that
    contains the image `ID` and `CLASS` prediction for the images listed
    in the **test.csv** file and submit it using the
    **`Submit Predictions`** tab.
4.  When you submit your file, the Public Leaderboard reflects the score
    for only **100% of the test set.**
5.  For generating your **`.csv`** file, use the `sample_submission.csv`
    file provided as reference.

You need to create a model that predicts the category of sports
represented in each image.

**Note: The entire dataset was first mixed and then randomly split into
train and test sets. You can split the public train set into train and
validation sets yourself.**

------------------------------------------------------------------------

### 2.1 File Descriptions<a
href="https://courses.opencv.org/api/jupyter/render_notebook/?url=https%3A%2F%2Fwww.dropbox.com%2Fscl%2Ffi%2F00njxjfemym5pr09yfe7k%2FProject2_Overview_Round3.ipynb%3Frlkey%3Dl3u79hzswuml2jcar9axxzgkr%26st%3Ditqw3qel%26dl%3D1&amp;images_url=#2.1-File-Descriptions"
class="anchor-link" target="_blank">¶</a>

**train.csv** - the training set, contains image `ids` with
corresponding `labels`. **test.csv** - the test set, contains image
`ids` **sample_submission.csv** - a sample submission file in the
correct format **images** - the image files (each image is named with
its corresponding id)

**Note: The training set (train.csv) contains image ids with
corresponding labels, whereas the test set (test.csv), contains only
image ids. So you cannot use the test set for validation. But you can
split the training set up into training and validation sets.**
